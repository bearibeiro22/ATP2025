 2b) Lista formada pelas palavras do texto iniciadas por maiúscula

```python
texto = """Vivia há já não poucos anos algures num concelho do Ribatejo um pequeno lavrador e negociante de gado chamado Manuel Peres Vigário"""
lista = [p for p in texto.split() if p[0].isupper()]
print(lista)
```

 3a) Conta os elementos de uma lista (sem usar len):

```python
# usando função auxiliar (não é necessário)
def meuLen_aux(lista):
    cont=0
    return meuLen(lista,cont=0)

def meuLen(lista,cont=0):
    if lista==[]:
        return cont
    else:
        return meuLen(lista[1:],cont+1)

meuLen([1,2,3])
```

```python
# forma mais simples
def meuLen(lista):
    if lista == []:
        return 0
    else:
        return 1 + meuLen(lista[1:])

print(meuLen([1, 2, 3]))
```

3b) Especifique uma função que calcula a potência de uma base 
levantada a um expoente (sem usar o operador **):

```python
def pot(base,expoente):
    if expoente==0:
        return 1
    else:
        return base*pot(base,expoente-1)

pot(2,6)
```

3c) Especifique uma função que recebe uma string e substitui 
todas as vogais por '*'

```python
def codAux(s,s_final):
    if s=="":
        return s_final
    
    elif s[0] in "aeiouAEIOU":
            s_final=s_final+"*"
            return codAux(s[1:], s_final)
    
    else:
        s_final=s_final+s[0]
        return codAux(s[1:], s_final)
    
    return s_final


def codVogais(s):
    s_final=""
    return codAux(s,s_final)


print(codVogais('Hoje ca'))
```
