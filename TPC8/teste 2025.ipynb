a) contaEspacos: defina uma função que recebe uma string e devolve o número de espaços na string.


```python
def contaEspacos(string):
    string = string.split()
    espacos = len(string) - 1
    return espacos

string = "Eu não gosto nada de estudar"
print(contaEspacos(string))
```

```python
# IMPORTANTE !!!
def contaEspacos_rec(string):
    if string=="":
        return 0
    if string[0] == " ": # está a ensinar que quando houver espaço é para contar +1
        return 1 + contaEspacos(string[1:])
    else:
        return contaEspacos(string[1:])


string = "Eu não gosto nada de estudar"
print(contaEspacos_rec(string))
```

b) ordenada?: defina uma função que recebe uma lista como argumento e devolve o valor booleano
verdadeiro se a lista estiver ordenada e falso caso contrário.

```python
def ordenada(lista):
    ordenada=True
    i=0
    while i<len(lista)-1 and ordenada:
        if lista[i]<=lista[i+1]:
            i=i+1
        else:
            ordenada=False
    return ordenada


lista1=[1,1,2,3,4]
lista2=[2,1,3,4]
print(ordenada(lista1))
print(ordenada(lista2))
```

```python
# IMPORTANTE !!!

def ordenada_rec(lista):
    if len(lista)<=1:
        return True
    if lista[0]<=lista[1]:
            return ordenada_rec(lista[1:]) # para começarmos a comparar o 1 com o 2 (isto basicamente vai "eliminar" o primeiro termo da lista quando recursa)
    else:
        return False

lista1=[1,1,2,3,4]
lista2=[2,1,3,4]
print(ordenada_rec(lista1))
print(ordenada_rec(lista2))
```

c) existeElem: defina uma função que recebe uma lista e um elemento e devolve 
True
 ou 
False

elemento exista ou não na lista.

```python
def existeElem(lista,elem):
    existe=False
    i=0
    while i<len(lista) and existe==False:
        if lista[i]==elem:
            existe=True
        i=i+1
    return existe

lista1=[1,2,3,4]
lista2=[1,2,3,4]
print(existeElem(lista1,3))
print(existeElem(lista2,5))
```

```python
# IMPORTANTE !!!
def existeElem_rec(lista,elem):
    if lista==[]:
        return False
    if lista[0]==elem:
        return True
    else:
        return existeElem_rec(lista[1:],elem)

lista1=[1,2,3,4]
lista2=[1,2,3,4]
print(existeElem_rec(lista1,3))
print(existeElem_rec(lista2,5))
```

```python
def existeElem_rec2(lista, elem, i=0):
    if i >= len(lista):
        return False
    if lista[i] == elem:
        return True
    return existeElem_rec2(lista, elem, i+1)

lista1=[1,2,3,4]
lista2=[1,2,3,4]
print(existeElem_rec2(lista1,3))
print(existeElem_rec2(lista2,5))
```

N
, e calcula o maior número de um digito (1 a 9) que divide cada número de entre

. O resultado será um dicionário.


In:  print(maiorDivisor(20) 

Out: {  1: 1,

    ...

    ... 
18: 9,

```python
# IMPORTANTE !!!
def maiorDivisor(num):
    res={}
    for i in range(1,num+1):
        divisores=[j for j in range(1,i) if i%j==0]
        if divisores: # significa se existir uma lista (divisores==True)
            res[i]=max(divisores) # retorna o máximo
        else:
            res[i]=1

    return res

print(maiorDivisor(20))
print(maiorDivisor(19))
```

 mmc: defina uma função que calcula o mínimo múltiplo comum entre dois números inteiros.

```python
def mdc(a,b):
    if a<b:
        a,b=b,a
    while a%b!=0:
        a,b=b,a%b
    return b

def mmc(a, b):
    return (a * b) // mdc(a, b)

print(mdc(3,5))
print(mmc(3,5))
print(mdc(4,2))
print(mmc(4,2))
print(mmc(4, 6))
print(mmc(12, 16))
```

d) reduxInt: defina uma função que recebe um número inteiro positivo como argumento e vai adicionar
repetidamente os seu dígitos até obter apenas um dígito que é retornado como resultado.

```python
def reduxInt(num):
    while num>=10:
        soma=0
        for digito in str(num):
            soma=soma+int(digito)
        num=soma
    return num

print(reduxInt(998))
```

f) ocupar_cama: Defina uma função que recebe um hospital, o nome de uma unidade hospitalar e o
identificador de uma cama, e caso a cama esteja livre devolve um novo hospital com a cama ocupada (não
faz nada se a cama estiver ocupada).

```python
def ocupar_cama(hospital, unidade, id):
    i = 0
    while i < len(hospital):
        if hospital[i][0] == unidade:
            j = 0
            while j < len(hospital[i][2]):
                cama = hospital[i][2][j]
                if cama[0] == id and cama[1] == False:
                    hospital[i][2][j] = (id, True)
                j = j + 1
        i = i + 1
    return hospital


hospital = [ ( "Unidade de Cardiologia", [ (1, "Dr. João"), (2, "Dra. Maria")], [(101, False), (102, True), (103, False), (104, False)] ), ( "Unidade de Pediatria", [ (3, "Dr. Pedro"), (4, "Dra. Clara") ], [ (201, True), (202, True), (203, True)] ) ] 
print(ocupar_cama(hospital,"Unidade de Cardiologia",101))
print(ocupar_cama(hospital,"Unidade de Cardiologia",102))
```

j) unidades_ocupacao_critica: Defina uma função que recebe um hospital e devolve uma lista de tuplos
(pares), em que o primeiro elemento do tuplo é o nome de uma unidade hospitalar e o segundo é a
percentagem de camas ocupadas nessa unidade, mas apenas deverão constar tuplos em que a
percentagem é maior ou igual a 80%.

```python
def unidades_ocupacao_critica(hospital):
    lista=[]
    for unidade in hospital:
        ocupadas=0
        ncamas=0
        for camas in unidade[2]:
            ncamas=ncamas+1
            if camas[1]==True:
                ocupadas=ocupadas+1
        percentagem=(ocupadas/ncamas)*100
        if percentagem>=80:
            lista.append((unidade[0],percentagem))
    return lista


hospital = [ ( "Unidade de Cardiologia", [ (1, "Dr. João"), (2, "Dra. Maria")], [(101, False), (102, True), (103, False), (104, False)] ), ( "Unidade de Pediatria", [ (3, "Dr. Pedro"), (4, "Dra. Clara") ], [ (201, True), (202, True), (203, True)] ) ] 
print(unidades_ocupacao_critica(hospital))
```
